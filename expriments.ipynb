## Function that transform columns contains dictionaries as values into columns

```python
import ast  # For safely evaluating the string dictionaries


def dict_to_column(df, list_features):
    """ Methode to handle columns with values as a dictionnary"""
    
    list_dataframe = []
    for column in list_features:
        # Create a DataFrame
        #df = pd.DataFrame(df_result)

        # Convert the string representations of dictionaries to actual dictionaries using ast.literal_eval
        df[column] = df[column].apply(ast.literal_eval)

        # Normalize the 'features' column to separate keys into columns
        df_attributes = pd.json_normalize(df[column])

        # Combine the new columns with the original DataFrame
        data = pd.concat([df.drop(columns=[column]), df_attributes], axis=1)
        list_dataframe.append(data) 

    all_df = pd.concat([list_dataframe[0], list_dataframe[1]], axis=1) 

    # remove duplacated columns
    all_df = all_df.loc[~all_df.index.duplicated(),:].copy()
    all_df = all_df.loc[:,~all_df.apply(lambda x: x.duplicated(),axis=1).all()].copy()
    all_df = all_df.drop(["feat_infos_dic1", "feat_infos_dic2"], axis=1)

    return all_df

 
#data = dict_to_column(df, list_features=["feat_infos_dic1", "feat_infos_dic2"])
```

```python
import ast  # For safely evaluating the string dictionaries


def to_string(value):
    return str(value) 

def dict_to_column(df, list_features):
    """ Methode to handle columns with values as a dictionnary """
       
    list_dataframe = []
    for column in list_features: 
        
        #Convert dict to string
        df2[column]= df2[column].apply(to_string)
        # Convert the string representations of dictionaries to actual dictionaries using ast.literal_eval
        df[column] = df[column].apply(ast.literal_eval)

        # Normalize the 'features' column to separate keys into columns
        df_attributes = pd.json_normalize(df[column])

        # Combine the new columns with the original DataFrame
        data = pd.concat([df.drop(columns=[column]), df_attributes], axis=1)
        list_dataframe.append(data) 

    all_df = pd.concat([list_dataframe[0], list_dataframe[1]], axis=1) 

    # remove duplacated columns
    all_df = all_df.loc[~all_df.index.duplicated(),:].copy()
    all_df = all_df.loc[:,~all_df.apply(lambda x: x.duplicated(),axis=1).all()].copy()
    all_df = all_df.drop(["feat_infos_dic1", "feat_infos_dic2"], axis=1)

    return all_df

 
data = dict_to_column(df2, list_features=["dict2", "dict1"])

data.head(2)
```

```python
import re 


def clean_column(value):
    try:
        value = re.sub("[^0-9]","", value)#.replace("DH", "")
    except:
        value = "NaN"
    return value

df1["price"] = df1["price"].apply(clean_column)
df1["piece"] = df1["piece"].apply(clean_column)
df1["rooms"] = df1["rooms"].apply(clean_column)
df1["bathroom"] = df1["bathroom"].apply(clean_column) 
df1["surface"] = df1["surface"].apply(clean_column)   

df1.head(2)
```

```python


# format features
new_colums_name = [string.lower().replace(" ", "_") for string in df1.columns.to_list()]
dic_ond_new_features = {}
for old_name, new_name in zip(df1.columns, new_colums_name):
    dic_ond_new_features.update({old_name: new_name})


df1["étage_du_bien"] = df1["étage_du_bien"].apply(clean_column)

df1 = df1.rename(columns=dic_ond_new_features) 
df1.head()
```
