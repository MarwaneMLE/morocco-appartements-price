```python
def scrape_link(start_page, end_page):
    """ 
    Methode to scrape the link and shared date of real state
    """
    url_pages = [] 
    for i in range(start_page, end_page + 1):
        url = f"https://www.mubawab.ma/fr/ct/casablanca/immobilier-a-vendre:p:{i}"
        url_pages.append(url)

    shared_date = []
    list_urls = []
    for i in range(len(url_pages)):
        url_page = url_pages[i]

        driver = webdriver.Chrome(options=options)
        driver.get(url_page)
        time.sleep(1)

        # Récupère tous les éléments ayant la classe 'contentBox'
        appartements = driver.find_elements(By.CLASS_NAME, "contentBox")

        # Parcourt chaque élément trouvé
        for appartement in appartements:
            try:
                # Trouve le lien (élément <a>) à l'intérieur de l'élément avec la classe 'listingTit'
                lien_element = appartement.find_element(By.CLASS_NAME, "listingTit").find_element(By.TAG_NAME, "a")
                date = appartement.find_element(By.CLASS_NAME, "listingDetails").text
                
                # Si le lien est trouvé, récupère l'attribut href
                if lien_element is not None:
                    lien = lien_element.get_attribute("href")
                    #print(lien)
                    list_urls.append(lien)

                if date is not None:
                    shared_date.append(date)

            except Exception as e:
                # Si un élément n'est pas trouvé ou qu'il y a une erreur, on l'ignore et passe au suivant
                #print(f"Erreur : {e}")
                continue

        df = pd.DataFrame()
        df["shared_date"] = shared_date
        df["link"] = list_urls
        df.to_csv(f"link-file.csv", index=False)

    return df
```
